```python
import pandas as pd
import math
import numpy as np
from Collaborative_Filtering import Euclidian_Similarity, Recommendations
```

# 1.0 Load Data

```python
users_feedback = {
    'Mary': {'The Hobbit': 2.5, 
             'The Lord of the Rings': 3.5,
             'Star Trek': 3.0, 
             'The Terminator': 3.5, 
             'Norbit': 2.5, 
             'Star Wars': 3.0
            },
    'Peter': {'The Hobbit': 3.0, 
              'The Lord of the Rings': 3.5, 
              'Star Trek': 1.5, 
              'The Terminator': 5.0, 
              'Norbit': 3.0, 
              'Star Wars': 3.5
             }, 
    
    'Stuart': {'The Hobbit': 2.5, 
               'The Lord of the Rings': 3.0,
               'The Terminator': 3.5, 
               'Star Wars': 4.0
              },
    
    'Jessica': {'The Lord of the Rings': 3.5, 
                'Star Trek': 3.0,
                'The Terminator': 4.0, 
                'Norbit': 2.5, 
                'Star Wars': 4.5
               },
    
    'Paul': {'The Hobbit': 3.0, 
             'The Lord of the Rings': 4.0, 
             'Star Trek': 2.0, 
             'The Terminator': 3.0, 
             'Norbit': 2.0,
             'Star Wars': 3.0
            }, 
    
    'Suzane': {'The Hobbit': 3.0, 
               'The Lord of the Rings': 4.0,
               'The Terminator': 5.0, 
               'Norbit': 3.5, 
               'Star Wars': 3.0
              },
    
    'Fred': {'The Lord of the Rings':4.5,
             'The Terminator':4.0,
             'Norbit':1.0
            }
}
```

```python
df_ratings = pd.DataFrame(users_feedback)
df_ratings = df_ratings.T
df_ratings
```

# 2.0 Euclidian Similarity

### 2.1 User-Based

```python
df_ratings
```

```python
#display(df_ratings)        
user_eu_sim = Euclidian_Similarity(df_ratings, verbose=0)
user_eu_sim.display_euclidian_similarity_matrix()  # display the similarity scores matrix
user_eu_sim_df = user_eu_sim.get_sim_scores_df()  # get the similarity scores matrix
display(user_eu_sim_df)

# get sim scores for a specific user
user_fred_euSim = user_eu_sim.get_similarity_with('Fred')
display(user_fred_euSim)

# get sim scores for a specific pair of users
user_sim_scores = user_eu_sim.get_sim_scores_df()
print('\nFred <--> Mary\n',user_sim_scores.loc['Fred','Mary'].round(3))
```

### 2.2 Item-based Euclidian Similarity

```python
df_ratings.T
```

```python
item_eu_sim = Euclidian_Similarity(df_ratings.T, verbose=0)
item_eu_sim.display_euclidian_similarity_matrix()  # display the similarity scores matrix
item_eu_sim_df = item_eu_sim.get_sim_scores_df()  # get the similarity scores matrix
display(item_eu_sim_df)

# get sim scores for a specific movie
item_Norbit_euSim = item_eu_sim.get_similarity_with('Norbit')
display(item_Norbit_euSim)

# get sim scores for a specific pair of movies
item_sim_scores = item_eu_sim.get_sim_scores_df()
print('\nStar Trek <--> The Terminator\n',item_sim_scores.loc['Star Trek','The Terminator'].round(3))
```

# 3.0 Recommendations

### 3.1 User-Based: Given a user, recommend one or more Features

```python
df_ratings
```

```python
# create similarity matrix - user based
user_eu_sim = Euclidian_Similarity(df_ratings, verbose=0)
user_eu_sim_df = user_eu_sim.get_sim_scores_df()  # get the similarity scores matrix

# get recommendations
# __init__(self, obs, df_input, df_sim_scores, cf_type='user', verbose=0):
fred_rec = Recommendations('Fred', df_ratings, user_eu_sim_df, verbose=1)

threshold = 2.5
top_k=3
df_rec, rec_lst = fred_rec.get_recommendations(threshold=threshold, top_k=top_k)

print("\nWe set threshold: {}, top_k: {}".format(threshold, top_k))
print("And the recommendations are:")
display(rec_lst)
display(df_rec)
```

### 3.2 User-Based: Given a Feature, recommend one or more Users

```python
df_ratings.T
```

```python
# create similarity matrix - movie based
movie_based_df_ratings = df_ratings.T

movie_eu_sim = Euclidian_Similarity(movie_based_df_ratings, verbose=0)  # transpose the df
movie_eu_sim_df = movie_eu_sim.get_sim_scores_df()  # get the similarity scores matrix
print("displaying the similarity scores matrix")
display(movie_eu_sim_df)

print("getting the recommendations")
rec = Recommendations('Star Trek', movie_based_df_ratings, movie_eu_sim_df, verbose=0)

threshold = 2.5
top_k=3
df_rec, rec_lst = rec.get_recommendations(threshold=threshold, top_k=top_k)

print("\nWe set threshold: {}, top_k: {}".format(threshold, top_k))
print("And the recommendations are:")
display(rec_lst)
display(df_rec)
```
